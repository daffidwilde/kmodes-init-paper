```python
from kmodes.kmodes import init_huang, init_matching

import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../data/zoo.csv")
```

```python
df = df.drop(["animal_name", "class"], axis=1)
```

```python
X = df.values
k = 3
```

```python
def dissim(X, x):
    return np.sum(X != x, axis=1)
```

```python
np.random.seed(0)
huang = init_huang(X, k, dissim)
```

```python
np.random.seed(0)
match = init_matching(X, k, dissim, "matching_huang")
```

```python
np.where(np.all(huang == match, axis=1))
```
