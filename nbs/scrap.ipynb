```python
from kmodes.kmodes import KModes
import pandas as pd
```

```python
name = "breast_cancer"

data = pd.read_csv(f"../data/{name}.csv").dropna().drop("class", axis=1)
df = pd.read_csv(f"../data/nclasses/{name}_results.csv")
df.head()
```

```python
worsened = df[df["initial_cost"] < df["final_cost"]]
worsened
```

```python
for n_clusters, seed, initialisation in worsened[
    ["n_clusters", "seed", "initialisation"]
].values:
    km = KModes(n_clusters, init=initialisation, random_state=seed, n_init=1)
    km.fit(data)
    print(
        n_clusters,
        initialisation,
        seed,
        km.epoch_costs_[0],
        km.epoch_costs_[-1],
        km.n_iter_,
    )
```
