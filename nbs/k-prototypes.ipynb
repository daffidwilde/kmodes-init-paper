```python
from kmodes.kprototypes import KPrototypes
from sklearn.metrics import silhouette_score

import json

import numpy as np
import pandas as pd
```

```python
with open("/Volumes/thesis-data/dtype_dict.json") as f:
    dtypes = json.load(f)
```

```python
df = pd.read_csv(
    "/Volumes/thesis-data/main.csv",
    dtype=dtypes,
    parse_dates=["PERIOD", "ADMDATE", "DISCDATE"],
)
```

```python
df["PERIOD"]
```

```python
cat_cols = df.select_dtypes(["category", "object", "datetime"]).columns
num_cols = df.select_dtypes(["float", "int"]).columns

categorical = [i for i, c in enumerate(df.columns) if c in cat_cols]
```

```python
df[num_cols] = df[num_cols].fillna(df[num_cols].mean())
```

```python
km = KPrototypes(n_clusters=2)

km.fit_predict(df.values, categorical=categorical)
```
