```python
import pandas as pd
import numpy as np
import itertools
```

## Dissimilarity

```python
def dissim(X, x):
    return np.sum(X != x, axis=1)


def summed_dissim(X, x):
    return np.sum(dissim(X, x))
```

```python
df = pd.read_csv("../../data/toy_car_example.csv", dtype="category")
df.index += 1
```

```python
df.to_latex("../../tex/example-table.tex", column_format="ccccccc")
```

## Mode by theorem

```python
X = df.values
```

```python
top = df.describe().iloc[2].values
top
```

```python
mode_str = "\\[ \n "
mode_str += "\mu = \left[ "
for attr in top:
    mode_str += "\\text{" + attr + "}, \\ "
mode_str = mode_str[:-4]
mode_str += " \\right] \\ \n "
mode_str += "\\]"
```

```python
mode_str
```

```python
with open("../../tex/top-mode.tex", "w") as f:
    f.write(mode_str)
```

```python
price, costs, doors, passengers, wheels, eco = [df[col].unique() for col in df.columns]
total_attr_space = itertools.product(price, costs, doors, passengers, wheels, eco)

space = []
for point in total_attr_space:
    space.append(np.array(point))

summed_dissims = [summed_dissim(X, point) for point in space]
idx = np.argmin(summed_dissims)

np.all(space[idx] == top)
```
